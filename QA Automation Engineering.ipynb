```python
import psycopg2
import os
import subprocess
```

```python
# Open connection with DB.
# Create DB.

connect = psycopg2.connect(
  database="postgres", 
  user="postgres", 
  password="dataarmor", 
  host="127.0.0.1", 
  port="5432"
)
```

```python
# Create a Table in DB.

cursor = connect.cursor()

cursor.execute('''CREATE TABLE if not exists TEST  
	(COL1 INT,
	COL2 INT,
	COL3 VARCHAR(50),
	COL4 VARCHAR(50),
	COL5 TIMESTAMP);''')
```

```python
# Filling a table with arbitrary data.
# Save changes.

for i in range(1000):
	if i % 20  == 0:
		cursor.execute(
			"INSERT INTO TEST (COL1,COL2,COL3,COL4,COL5) \
				VALUES (12,  18, '', '','1999-01-08 22:10');"
		)
	elif i % 15 == 0:
		cursor.execute(
			"INSERT INTO TEST (COL1,COL2,COL3,COL4,COL5) \
				VALUES (NULL,  NULL, NULL, NULL, NULL);"
		)
	else:	
		cursor.execute(
			"INSERT INTO TEST (COL1,COL2,COL3,COL4,COL5) \
				VALUES (12,  18, 'John', 'Computer Science','1999-01-08 22:10');"
		)

connect.commit()
```

```python
# 1. First, we need to connect to the system and authorize in it. Note that session is active during 10 minutes since the
# last action.

subprocess.run('executecommand.bat connect -host 127.0.0.1 -port 11000 -login admin -password Education%21 -protocol https')
```

```python
# 2. Let's change the Audit Storage location from the SQLite database to a PostgreSQL. It is recommended if you
# expect a large volume of traffic to be transferred to the Audit Storage.

subprocess.run('executecommand.bat changeStorage -dbType postgresql -host 127.0.0.1 -database postgres \
                                            -schema public -password dataarmor -login postgres -port 5432')
```

```python
# 3. Register the database that is required to be monitored (the target database). We will use a MySQL database that is
# located at 192.168.1.71:3305. We need to open a proxy on a local interface port so that the database clients will be
# able to connect to 127.0.0.1:1025 like to the real database server.

subprocess.run('executecommand.bat addInstancePlus -dbType postgresql -dbHost 127.0.0.1 -dbPort 5432 \
        -proxyHost  0.0.0.0 -proxyPort 1025 -database postgres -password dataarmor -login postgres -name test')
```

```python
# 4. We need to create several auxiliary objects that will help us to filter according to the specified IP address and user
# name.

subprocess.run('executecommand.bat addHost -startIPv4 127.0.0.1 -name testhost -endIPv4 127.0.0.254')
subprocess.run('executecommand.bat addDbUser -name user_1 -dbType postgresql -instance test')
```

```python
# 5. In order to filter queries by code, we need to create a group of queries. We will add a SQL query that is required
# to be logged to this group of queries. A group can contain many SQL queries specified as regular expressions or as
# full text of a SQL query.

subprocess.run('executecommand.bat addQueryGroup -name test_postgresql_gr')
subprocess.run('executecommand.bat addQueryOfGroup -name test_postgresql_gr -sql ".*test.*" -regExp true')
```

```python
# 6. Sometimes it is necessary to restrict active period of a Rule. In that case you can use Schedules. To create a
# Schedule, use the addSchedule command.

subprocess.run('executecommand.bat addSchedule -name test_schedule -intervals mo08:00:00-18:00:00;tu08:00:00-18:00:00')
```

```python
# 7. Using all the objects specified above we create an audit Rule:

subprocess.run('executecommand.bat addAuditRule -action audit -instance test -dbType postgresql -filterType object \
                                                                -logData true -logInStorage true -name "AuditTestRule"')
```

```python
# 8. Create a mask Rule:

subprocess.run('executecommand.bat addMaskRule -instance test -login postgres -password dataarmor -dbType postgresql -logData true -logInStorage true -maskType maskFirst -maskCount 2 -maskColumns "test.col3;test.col4;" -enable true -name MaslRule')
```
